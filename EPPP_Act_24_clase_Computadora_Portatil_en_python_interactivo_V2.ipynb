```python
# Definición de la Clase
class ComputadoraPortatil:
    def __init__(self, marca, bateria, peso):
        self.marca = marca
        self.bateria = bateria  # Ejemplo: 100% o 5000mAh
        self.peso = peso        # Ejemplo: 1.5kg

    # --- Métodos (Funciones solicitadas) ---
    def leerdatos(self):
        """Este método simula la actualización o lectura de estado del hardware"""
        print(f"\n🔍 Analizando estado de la Laptop {self.marca}...")
        print(f"Verificando integridad del peso ({self.peso}) y celdas de batería.")

    def mostrardatos(self):
        """Muestra la información técnica del objeto"""
        print(f"\n💻 ESPECIFICACIONES TÉCNICAS")
        print(f"Marca: {self.marca}")
        print(f"Capacidad de Batería: {self.bateria}")
        print(f"Peso del dispositivo: {self.peso}")

# --- Función Principal ---
def sistema_computo():
    laptops = [] # Lista para guardar los dos objetos (laptop1, laptop2)

    # Registro de los 2 objetos solicitados
    for i in range(1, 3):
        print(f"\n--- REGISTRO DE LAPTOP {i} ---")
        m = input(f"Ingrese la marca de la laptop {i}: ")
        b = input(f"Ingrese capacidad de batería (ej. 10 hrs): ")
        p = input(f"Ingrese el peso (ej. 1.2 kg): ")

        # Creamos el objeto y lo guardamos en la lista
        nueva_laptop = ComputadoraPortatil(m, b, p)
        laptops.append(nueva_laptop)

    # Menú Interactivo
    while True:
        print("\n" + "═"*35)
        print("     SISTEMA DE GESTIÓN TÉCNICA")
        print("═"*35)
        print(f"1. Gestionar {laptops[0].marca}")
        print(f"2. Gestionar {laptops[1].marca}")
        print("3. Salir")

        opc = input("\n¿Qué laptop desea consultar? (1-3): ")

        if opc in ["1", "2"]:
            # Elegimos el objeto según el índice de la lista
            lap = laptops[int(opc) - 1]

            print(f"\n--- Acciones para: {lap.marca} ---")
            print("a) Leer Datos (Diagnóstico)")
            print("b) Mostrar Datos (Especificaciones)")
            accion = input("Selecciona acción (a o b): ").lower()

            if accion == "a":
                lap.leerdatos()
            elif accion == "b":
                lap.mostrardatos()
            else:
                print("Acción no válida.")

        elif opc == "3":
            print("\nApagando sistema de gestión...")
            break
        else:
            print("\n[!] Opción no válida.")

# --- Ejecución ---
if __name__ == "__main__":
    sistema_computo()

    print("\n" + "═"*45)
    print("Programa terminado con éxito")
    print("Desarrollado siguiendo el modelo de Alejandra Rivera Valenzuela")
    print("═"*45)
```
